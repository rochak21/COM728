```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
import tui_ 
import retrieve #custome module for data retrieval funtions
import analyse #custom module for data analysis functions
import visualise #custom module for data visualisation functions

def user_input(user_prompt, valid_choices):
    
    while True:
        choice = input(user_prompt).lower()
        if choice in valid_choices:
            return choice
        else:
            tui_.invalid_choice() #calls function from user_interface module to handle invalid user inputs. 
            
def display_main_menu():
    """
    Function to display the main menu.
    This function prints options available in main menu.
    """
    print("--- Main Menu ---") 
    print(" ")
    print("1. Retrieve Data")
    print("2. Analyse Data")
    print("3. Data Visualisation")
    print("5. Exit")
    print(" ")

def read_sub_menu():
    print(" ")    
    print("--- Data Retrieval ---")
    print("a1. Retrieve data based on the Employee Number")
    print("a2. Retrieve data based on the Job Role")
    print("a3. Retrieve data based on the Department")
    print("a4. Retrieve data based on the Education level")
    print("5. Return to Main Menu")
    print(" ")
    
def analyse_sub_menu():
    print(" ")    
    print("--- Analyse Data ---")
    print("b1. Identify data based on the education level")
    print("b2. Analyse avg monthly rate based on specific education level")
    print("b3. Analyse avg duartion of employee in a specific department")
    print("b4. Analyse data based on the Business Travel")
    print("5. Return to Main Menu")
    print(" ")    
def visualise_sub_menu():
    print(" ")    
    print("--- Data Visualisation ---")
    print("c1. Create a chart to show the proprotion of employee based on education levels")
    print("c2. Create a chart to compare the frequency of employee training indaily each education field")
    print("c3. Plot job satisfaction scores by job level for each department in a single chart")
    print("c4. Create a chat to compare monthly income based on education level")
    print("5. Return to Main Menu")
    print(" ")
    
def handle_data_read(data, tui_):
    """
    Function to handle data reading process based on the users input.
    
    Parameters:
    data (dict): the data set in dictionary format
    tui_ (module): used to interact with the user in specified format
    
    This function repeatedly displays the data read menu and executes the corresponding
    data retrieval function based on user input until the user decides to exit.
    """
    while True:
        read_sub_menu()
        choice = tui_.get_user_choice("Enter your choice from the menu [a1-a4, 5]: ", ['a1', 'a2', 'a3', 'a4', '5'])
        print()
        if choice == 'a1':
            result = retrieve.retrieve_by_demographic_value(data)
            if result:
                print("Gender: {}, Age: {}, Marital Status: {}, Department: {}".format(*result))
            else:
                print("Employee number not found.")
        elif choice == 'a2':
            result = retrieve.retrieve_by_job_role(data)
            if result:
                print("Department: {}, Education Level: {}, Total Working Years: {}, Standard Hours: {}".format(*result))
            else:
                print("Job role not found.")
        elif choice == 'a3':
            result = retrieve.retrieve_by_working_hours(data)
            if result:
                print("Employee Number: {}, Job Role: {}, Marital Status: {}, Hourly Rate: {}".format(*result))
            else:
                print("No employees found for the given department with Standard Hours less than 60.")
        elif choice == 'a4':
            result = retrieve.retrieve_job_role_info(data)
            if result:
                print("Business Travel: {}, Department: {}, Gender: {}, Education Field: {}".format(*result))
            else:
                print("Education level not found.")
        elif choice == '5':
            break
        input("Press Enter to return to the menu...") 
        print("")
def handle_analyse_data(df_data):
    while True:
        analyse_sub_menu()
        choice = tui_.get_user_choice("Enter your choice from the menu [b1-b5, 5]: ", ['b1', 'b2', 'b3', 'b4', 'b5', '5'])
        print()
        if choice == 'b1':
            analyse.analyse_by_job_role(df_data)
        elif choice == 'b2':
            analyse.analyse_by_education_level(df_data)
        elif choice == 'b3':
            analyse.analyse_by_department(df_data)
        elif choice == 'b4':
            analyse.analyse_by_education_field(df_data)
        elif choice == '5':
            break 
        input("Press Enter to return to the menu...")
        input(" ")   
def handle_data_visualisation(df_data):
    while True:
        visualise_sub_menu()
        choice = tui_.get_user_choice("Enter your choice from the menu [c1-c4, 5]: ", ['c1', 'c2', 'c3', 'c4', '5'])
        print()

        if choice == 'c1':
            visualise.visualise_education_proportion(df_data)
        elif choice == 'c2':
            visualise.visualise_training_frequency(df_data)  
        elif choice == 'c3':
            visualise.visualise_job_satisfaction(df_data)
        elif choice == 'c4':
            visualise.visualise_tenure_distribution(df_data)
        elif choice == '5':
            break

        input("Press Enter to return to the menu...")


def main():
    """
    Main function to run the program.
    """
    dict_data, df_data = tui_.load_csv_file() # Load data from a CSV file
    
    # Check if data is loaded successfully
    if dict_data is None or df_data is None:
        print("Exiting the program.")
        return  # Exit the program if data cannot be loaded
    
    #main loop to continuosly display main menu and handle user choices inputted
    while True:
       
        tui_.welcome()
        display_main_menu()
        choice = tui_.get_user_choice("Please enter your choice to enter the menu [1-3, 5]: ",['1','2','3','5'])
        #manages users input by calling tui functions
        
        if choice == '1':
            handle_data_read(dict_data, tui_)
        elif choice == '2':
            handle_analyse_data(df_data)
        elif choice == '3':
             handle_data_visualisation(df_data)
        elif choice == '5':
            break
            
#main function that is called only when script is direclty executed
if __name__ == "__main__":
    
    main()
```
